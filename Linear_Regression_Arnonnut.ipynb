```python
import matplotlib.pyplot as plt

x = [-0.1382643, -0.46947439,  0.76743473, -0.23413696, -0.23415337,  0.54256004,  0.49671415,  1.52302986,  1.57921282,  0.64768854,]
y = [-2.97603162, -8.88788838, 13.58267603, -5.01913245, -3.84927043,  8.79528589, 8.74581338, 28.16767021, 28.50000881, 10.47032021]

plt.scatter(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot of x and y')
plt.show()
```

```python
# prompt: Use linear regression(i.e. Scikit-lean) to draw linear line between x and y data

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Convert lists to numpy arrays
x = np.array(x).reshape(-1, 1)
y = np.array(y)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(x, y)

# Get the slope and intercept of the line
slope = model.coef_[0]
intercept = model.intercept_

# Print the equation of the line
print("Equation of the line: y = {:.2f}x + {:.2f}".format(slope, intercept))

# Predict y values for the given x values
y_pred = model.predict(x)

# Plot the data and the linear regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Linear Regression')
plt.show()
```

```python
# prompt: assume that we have data
# time (x) = [5, 7, 12, 6, 30]
# mass (y) = [40, 120, 180, 210, 240]
# perform follwing
# - plot the data
# - Plot the linear regression with SK-lean
# - plot te linear model
# - find the linear coefficients a,b (y=ax+b)

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
time = np.array([5, 7, 12, 16, 30]).reshape(-1, 1)
mass = np.array([40, 120, 180, 210, 240])

# Create linear regression object
model = LinearRegression()

# Fit the model
model.fit(time, mass)

# Make predictions
mass_pred = model.predict(time)

# Plot the data and the linear regression line
plt.scatter(time, mass, label='Data')
plt.plot(time, mass_pred, color='red', label='Linear Regression')
plt.xlabel('Time')
plt.ylabel('Mass')
plt.title('Linear Regression of Mass vs. Time')
plt.legend()
plt.show()

# Get the coefficients
a = model.coef_[0]
b = model.intercept_

print("Linear coefficients:")
print("a (slope) =", a)
print("b (intercept) =", b)
```

#HW 5.6

Find Linear regression that fit following data by assume the order of  

$𝑦=𝜃_0+𝜃_1 𝑥+𝜃_2 𝑥^2+𝜃_3 𝑥^3$


* Plot the data

* Plot the regession


image.png

```python
# prompt: x = [3.1818, -3.5859, -4.6970, 4.4949, -1.4646, -1.8687, -2.1717, -3.2828, -3.6869, 3.6869]
# y = [0.4375, -251.5471, -517.8529, 66.7037, -25.7354, -39.7799, -73.4630, -208.4632, -279.0183, 8.1043]
# Find Linear regression that fit following data by assume the order of
# y=θ0+θ1x+θ2x2+θ3x3
# Plot the data
# Plot the regession

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x = [3.1818, -3.5859, -4.6970, 4.4949, -1.4646, -1.8687, -2.1717, -3.2828, -3.6869, 3.6869]
y = [0.4375, -251.5471, -517.8529, 66.7037, -25.7354, -39.7799, -73.4630, -208.4632, -279.0183, 8.1043]

x = np.array(x).reshape(-1, 1)
x_2 = x**2
x_3 = x**3

X = np.concatenate((x, x_2, x_3), axis=1)
y = np.array(y)

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.scatter(x, y, label='Data')
plt.plot(x, y_pred, color='red', label='Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression')
plt.show()

r2 = r2_score(y, y_pred)
print("R-squared:", r2)
```

#HW 5.7

create regression data using parameters: n_samples=30,
                       n_features=7, noise=10, n_informative=2,
                       effective_rank=3, random_state=11

compare the R2 values between standard linear regression and ridge regression using $λ = 0.1$


```python
# prompt: create regression data using parameters: n_samples=30, n_features=7, noise=10, n_informative=2, effective_rank=3, random_state=11
# compare the R2 values between standard linear regression and ridge regression using  λ=0.1

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import r2_score

# Generate regression data
X, y = make_regression(n_samples=30, n_features=7, noise=10, 
                       n_informative=2, effective_rank=3, random_state=11)

# Standard Linear Regression
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)
r2_linear = r2_score(y, y_pred_linear)
print("Standard Linear Regression R-squared:", r2_linear)

# Ridge Regression
ridge_model = Ridge(alpha=0.1)  # lambda is represented as alpha in sklearn
ridge_model.fit(X, y)
y_pred_ridge = ridge_model.predict(X)
r2_ridge = r2_score(y, y_pred_ridge)
print("Ridge Regression R-squared:", r2_ridge)
```

```python
# Step 1
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pandas as pd

data = pd.read_excel("salary_dataset.xlsx")
with open('salary_dataset.xlsx', 'r') as f:
  print(f)
print(data)
```

```python
# Step 2
# ตรวจสอบค่าที่หายไปในคอลัมน์ salary
salary_missing = pd.isna(data['salary'])
print("missing values in 'salary' column:", salary_missing.sum())

# ตรวจสอบค่าที่หายไปในคอลัมน์ experience
experience_missing = pd.isna(data['experience'])
print("missing values in 'experience' column:", experience_missing.sum())

# หา index ของค่าที่หายไปในคอลัมน์ salary
salary_missing_index = data[salary_missing].index.tolist()
print("missing values in 'salary':", salary_missing_index)

# หา index ของค่าที่หายไปในคอลัมน์ experience
experience_missing_index = data[experience_missing].index.tolist()
print("missing values in 'experience':", experience_missing_index)
```

```python
index_salary_empty = np.isnan(data.iloc[:, 5])
index_experience_empty = np.isnan(data.iloc[:, 1])

# รวม index ที่มีค่าหายไปในทั้งสองคอลัมน์โดยใช้ Logic AND
combined_index_AND = index_salary_empty & index_experience_empty

# ใช้ numpy.where() เพื่อตรวจสอบ index ที่มีค่าหายไป
rows_with_missing_values_AND = np.where(combined_index_AND)[0]
print("Index of rows with missing values:", rows_with_missing_values_AND)
```

```python
index_salary_empty = np.isnan(data.iloc[:, 5])
index_experience_empty = np.isnan(data.iloc[:, 1])

# รวม index ที่มีค่าหายไปในทั้งสองคอลัมน์โดยใช้ Logic OR
combined_index_OR = index_salary_empty | index_experience_empty

# ใช้ numpy.where() เพื่อตรวจสอบ index ที่มีค่าหายไป
rows_with_missing_values_OR = np.where(combined_index_OR)[0]

print("Index of rows with missing values:", rows_with_missing_values_OR)
```

```python
# ทำความสะอาดข้อมูลโดยการลบแถวที่มีค่าที่หายไป
data_cleaned = data.dropna()
print(data_cleaned)
```

```python
# Step 3
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
# Scatter plot for salary vs experience
plt.figure(figsize=(8, 6))
plt.scatter(data_cleaned['experience'], data_cleaned['salary'], color='blue', marker='x')
plt.xlabel('Experience')
plt.ylabel('Salary (baht)')
plt.title('Salary vs Experience')

# Define specific y-ticks and format them
plt.yticks(np.arange(10000, 55001, 5000))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x/10000:.1f}'.rstrip('0').rstrip('.')))

# Add 'x10^4' at the top-right of the plot
plt.text(plt.gca().get_xlim()[0], plt.gca().get_ylim()[1], r'$\times 10^{4}$', fontsize=12, verticalalignment='bottom', horizontalalignment='left')

plt.show()
```

```python
# Step 4
# Prepare the data
X = data_cleaned['experience'].values.reshape(-1, 1)
y = data_cleaned['salary'].values

# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the coefficients
w0 = model.intercept_
w1 = model.coef_[0]

print("Intercept (w0):", w0)
print("Coefficient (w1):", w1)

# Objective function of Linear Regression
print("\nObjective Function of Linear Regression: Minimize 1/2 * sum((y_pred - y_true)^2)")
```

```python
# Step 5 
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# Update the plot to match the reference image
plt.figure(figsize=(8, 6))
plt.scatter(data_cleaned['experience'], data_cleaned['salary'], marker='x', color='blue', label='Observations')
plt.plot(X, model.predict(X), color='red', label='Prediction')

# Set axis labels and title to match the reference image
plt.xlabel('working experience (year)')
plt.ylabel('Salary (baht)')
plt.title('Salary vs Experience (year) with Prediction')

# Define specific y-ticks and format them
plt.yticks(np.arange(10000, 55001, 5000))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x/10000:.1f}'.rstrip('0').rstrip('.')))

# Add 'x10^4' at the top-right of the plot
plt.text(plt.gca().get_xlim()[0], plt.gca().get_ylim()[1], r'$\times 10^{4}$', fontsize=12, verticalalignment='bottom', horizontalalignment='left')

# Set legend to match the reference image
plt.legend(loc='lower right')

# Show the updated plot
plt.show()
```
