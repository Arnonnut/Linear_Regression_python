## EX 6.1 Linear regression
image.png

````python
from sklearn import preprocessing
X = [(63,1.6),(72,1.83),(84,1.65),(67,1.57),(75,1.68)]
X.sort()
weight=[]
for i in X:
  weight.append(i[0])
height=[]
for i in X:
  height.append(i[1])

x_bar=sum(weight)/len(weight)
y_bar=sum(height)/len(height)

nomi=[]
for i in range(len(weight)):
  nomi.append((weight[i]-x_bar)*(height[i]-y_bar))

denom=[]
for i in weight:
  denom.append((i-x_bar)**2)

b=sum(nomi)/sum(denom)

a=y_bar-b*x_bar


print("a = ",a)
print("b = ",b)
````

## EX6.2 Linear Regression with LSM
image.png

````python
import numpy as np
import matplotlib.pyplot as plt
Y = np.array([[63],[72],[84],[67],[75]])
X =np.array([[1.6],[1.83],[1.65],[1.57],[1.68]])


A=np.column_stack(((np.ones(len(X))),np.array(X),np.array(X)**2))

p_inv=np.linalg.inv(np.dot(A.T,A))
theta=np.dot(np.dot(p_inv, A.T),Y)
print(theta)

y_plot=[]
for i in sorted(X):
  y_plot.append(theta[0]+theta[1]*i+theta[2]*(i**2))

plt.plot(sorted(X),y_plot)
plt.figure(figsize=(8, 6))

plt.plot(sorted(X), y_plot, label="Fitted Polynomial Curve", color='blue')
plt.scatter(X, Y, color='red', label="Data Points")
plt.title("Polynomial Fit of Weight vs. Height (Using Manual Calculation)")
plt.xlabel("Height (m)")
plt.ylabel("Weight (Kgs)")
plt.legend()
plt.grid(True)
plt.show()
````

## EX6.3 Logistic Regression
Find the Logistic Regression for following data
image.png

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn import preprocessing

# Generate synthetic data for binary classification
X = np.array([[63,1.6],[72,1.83],[84,1.65],[67,1.57],[75,1.68]])
y = np.array([[0],[0],[1],[1],[1]])
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)
# Create and train a logistic regression model
logistic_reg = LogisticRegression()
logistic_reg.fit(X, y)

# Make predictions on the test set
y_pred = logistic_reg.predict(X)

# Plot decision boundary (for demonstration purposes)
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.01), np.arange(x2_min, x2_max, 0.01))
Z = logistic_reg.predict(np.c_[xx1.ravel(), xx2.ravel()])
Z = Z.reshape(xx1.shape)

plt.contourf(xx1, xx2, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c="blue")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Logistic Regression Decision Boundary")
plt.show()
````

## Ex6.4 Predict value from Logistic Regression

Predict the status of weight = 67 , Height = 1.65 from model in 9.3

````python
new_data = np.array([[67, 1.65]])
scaled_data = scaler.transform(new_data)
prediction = logistic_reg.predict(scaled_data)
print(prediction[0])

if prediction[0] == 1:
    print("obese")
else:
    print("normal")
````
